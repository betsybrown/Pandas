# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name"])
#school_data_complete.count()
```

```python
#total schools
unique_schools = len(school_data_complete['school_name'].unique())
#unique_schools
```

```python
#total students
total_students = school_data_complete ['Student ID'].count()
#total_students
```

```python
#total budget
total_budget = school_data['budget'].sum()
#total_budget
```

```python
#average math scores
avg_math_score = (school_data_complete['math_score'].sum()/total_students).round(2)
#avg_math_score.round(2)
```

```python
#average reading scores
avg_reading_score = (school_data_complete['reading_score'].sum()/total_students).round(2)
#avg_reading_score.round(2)
```

```python
#combined math and reading average scores
avg_passing_rate = (avg_math_score.round(2)+avg_reading_score.round(2))/2
#avg_passing_rate
```

```python
#passing math 1 of 2
over70 = (student_data['math_score']>=70).sum()
#over70
```

```python
#passing math 2 of 2
over70passing = (over70*100/total_students).round(2)
#over70passing
```

```python
#passing reading 1 of 2
over70r = (student_data['reading_score']>=70).sum()
#over70r
```

```python
#passing reading 2 of 2
over70rpassing = (over70r*100/total_students).round(2)
#over70rpassing
```

```python
#overall passing
combined_average_passing = (((over70passing)+(over70rpassing))/2).round(2)
#combined_average_passing
```

```python
#create dataframe
district_summary = {
                                  "Total Schools": [unique_schools],
                                  "Total Students": [total_students],
                                  "Total Budget $": [total_budget],
                                  "Average Math Score": [avg_math_score],
                                  "Average Reading Score": [avg_reading_score],
                                  "% Passing Math": [over70passing],
                                  "% Passing Reading": [over70rpassing],
                                  "Overall Passing Rate": [combined_average_passing]
                                  
                                                          }
#district_summary
                                                          
```

```python
math_passing=student_data.loc[student_data['math_score']>=70,:]
#math_passing.count()
```

```python
total_math_passing=math_passing.groupby(['school_name']).count()['math_score']
#total_math_passing.sum()
```

```python
percent_math_passing = (total_math_passing/students_by_school)*100
#percent_math_passing
```

```python
reading_passing=student_data.loc[student_data['reading_score']>=70,:]
#reading_passing.count()
```

```python
total_reading_passing=math_passing.groupby(['school_name']).count()['reading_score']
#total_reading_passing.sum()
```

```python
percent_reading_passing = (total_reading_passing/students_by_school)*100
#percent_reading_passing
```

```python
overall_passing = (percent_math_passing+percent_reading_passing)/2
#overall_passing
```

```python
#create dataframe
school_summary = pd.DataFrame({   "School Type": sn_school_data,
                                  "Total Students": students_by_school,
                                  "Total Budget": budget_by_school,
                                  "Total Budget Per Student": per_student_budget_by_school.astype(int),
                                  "Average Math Score": avg_math_by_school,
                                  "Average Reading Score": avg_reading_by_school,
                                  "% Passing Math": percent_math_passing,
                                  "% Passing Reading": percent_reading_passing,
                                  "Overall Passing Rate": overall_passing
                                  
                                                          })
#school_summary
```
